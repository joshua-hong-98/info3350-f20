# Poisson distributions

Show the effects of varying `mu` of the Poisson discrete distribution.

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
fig, axs = plt.subplots(6, figsize=(4,12), sharey=True)
bins = list(range(18,90,1))
for i, mu in enumerate([1,2,5,10,30,50]):
    data = stats.poisson.rvs(loc=18, mu=mu, size=150000) # left edge at 18, average at mu
    axs.flat[i].hist(data,bins=bins);
```
