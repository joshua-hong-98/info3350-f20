# Cosine and sine values between 0 and 90 degrees

```python
import numpy as np
from matplotlib import pyplot as plt
vals = np.linspace(0,np.pi/2, 100)
cosines = [np.cos(val) for val in vals]
sines = [1-np.sin(val) for val in vals]
sq_sines = [np.sqrt(1-np.sin(val)**2) for val in vals]
fig, ax = plt.subplots(figsize=(12,8))
plt.plot(vals, cosines, c='k', label='cos(x)')
plt.plot(vals, sines, label='1 - sin(x)')
plt.plot(vals, sq_sines, '.', label='sqrt(1 - sin(x)^2)', alpha=0.5)
plt.xlabel('Angle (radians)')
plt.legend()
plt.savefig('cos-sin.png')
plt.show()
```
