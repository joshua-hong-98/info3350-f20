# Cosine and sine distances between 0 and 90 degrees

```python
import numpy as np
from matplotlib import pyplot as plt
vals = np.linspace(0,np.pi/2, 100)
cos_dist= [1-np.cos(val) for val in vals]
sin_dist = [np.sin(val) for val in vals]
fig, ax = plt.subplots(figsize=(12,8))
plt.plot(vals, cos_dist, label='Cosine distance')
plt.plot(vals, sin_dist, label='Sine distance')
plt.xlabel('Angle (radians)')
plt.title("Distance metrics")
plt.legend()
plt.savefig('cos-sin.png')
plt.show()
```
