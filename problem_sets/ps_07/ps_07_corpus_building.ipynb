# Problem set 7: Corpus building (mini-project, step 2)

## Summary

Contribute to a literary corpus built by the whole class. Everyone will use this corpus for the last part of the mini-project.

## Details

Our problem sets so far have relied on corpora (novels, Old Bailey transcripts) that I have curated for you. For the mini-project, you will each contribute to a literary corpus that we will explore and analyze.

Your assignment is to select, clean, and document **two** novel-like volumes from [Project Gutenberg](http://www.gutenberg.org/) (or a similar source) and to check the work of two of your classmates (hence, you will check four volumes).

### The big picture

* By "novel-like," I mean a *book of narrative fiction*. 
    * So ... not poetry, not drama, not nonfiction. 
    * Short-story *collections* are OK, so long as they are either by a single author or are grouped around a single genre or theme; individual short stories are not OK. 
    * Each volume should contain at least 20,000 words.
* The volumes you select should be in English and should have been written after roughly 1600.
    * Translations into English from other languages are OK. 
    * Volumes that retain archaic fetaures (of spelling, punctuation, ligatures, etc.) are not OK (don't worry, there aren't many of these in Gutenberg). 
    * Translations of ancient or classical texts are not OK, since the novel hadn't been invented at the time.
* One of your texts must be selected from the [horror](http://www.gutenberg.org/ebooks/bookshelf/42) (just in time for Halloween!) or [detective fiction](http://www.gutenberg.org/ebooks/bookshelf/30) bookshelves, or from an analgous bibliography of those genres.
    * The other text may be selected from any literary genre, including such non-genres as "classics" or "literature." (But see the note above about texts originally composed before 1600.)
* At least one of your texts must be by a non-cis-male author.
    * There aren't many trans or non-binary folks in Gutenberg, so this requirement mostly amounts to at least one woman. But any trans, non-binary, or other such author you can find is also OK.
    * Think about other ways in which you might diversify the corpus. 
        * See, for instance, Kenton and Howard Rambsy's ["300 African American Texts"](https://docs.google.com/spreadsheets/d/12lcDbFWMEzxulRteV6I7sEMI2WpZbg-mHlQeWPQoAm8/edit#gid=0) (and the accompanying [blog post](http://www.culturalfront.org/2020/09/beginning-with-dataset-of-300-african.html)).
        * Or try Gutenberg's shelf of [sci-fi titles by women](http://www.gutenberg.org/ebooks/bookshelf/403).
        * The library maintains good lists of [resources for literary scholarship](https://guides.library.cornell.edu/sb.php?subject_id=88071), some of which are bibliographies of postcolonial and minor literatures.
* Speaking of which, you are free to consult bibliographic resources outside Gutenberg to identify your texts and to gather metadata (see below). You are also free to use full texts sourced outside Gutenberg, so long as they are:
    * Legally obtained.
    * Produced and edited to a similar level of quality as are the Gutenberg texts.
    * **Avoid OCR!**
    
### The process

1. Record metadata for your texts in [this spreadsheet](https://docs.google.com/spreadsheets/d/1LHi_h4mrU8nK79vmErXOVC4EhphO0mBxsasgCTVMDf0/edit?usp=sharing). 
    * Enter your NetID in the `owner` column.
    * Insert rows in alphabetical order by author surname to reduce the chance of duplicates.
    * Note that you must be logged in to your Cornell Google account to access the spreadsheet (and the other resources used in this assignment).
1. Upload one UTF-8 encoded `.txt` file for each volume to [this Cornell Google Drive folder](https://drive.google.com/drive/folders/1lbeZiBAVCzjCWojCK8mfmELa-Q8FMNUm?usp=sharing).
    * You can take Gutenberg's word for the character encoding, or check it in Chrome (View -> Developer -> JavaScript Console -> `document.charset`) or Firefox (Tools -> Page info).
1. Check the work of two other students (four texts in sum). 
    * Do their files and metadata records meet all guidelines? 
    * Have they removed all boilerplate and any paratext that the class guidelines say should be removed?
    * In order to reduce the chance of local errors, you should check the work of students who did not check your own work.
    * If you find problems, email the woner with suggested fixes.
        * If you're the owner, it's your responsibility to fix any issues identified by the checkers.
    * After you have checked a text, enter your NetID in either the `check_1` or `check_2` field in the spreadsheet.

### Shared guidelines

Before we begin, we will decide how to edit the text files, which metadata to collect, and how to format everything. What analysis do we want to support, and what information do we need in order to carry out that analysis?

We'll record our decisions in this [shared Google document](https://docs.google.com/document/d/1GD1IpRsybW9R6uZWTOZB4aDizaJembbyS0nvU5dAPiA/edit?usp=sharing).


## Grading

You'll receive a simple completion grade for this problem set, similar to the 2, 1, 0 scoring for the response posts. 

* If you supply two texts as described above and check four texts, you'll receive a 2, which will translate to full credit for 10% of the next problem set. 
* If you do incomplete work (for example, supply only one text, supply poor-quality `.txt` files, don't check others' work, etc.), you'll get a 1, which will translate to 10% off your next problem set. 
* If you don't do any of the work, you'll get a **zero for the entire next problem set**.
